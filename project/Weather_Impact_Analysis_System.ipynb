```python
 # AgriClimate Solutions Limited

The Company develops software that helps farmers understand how daily weather conditions may affect crop production. As a Junior Python Developer, create a simple Python Program that summaries the weather data and highlighs conitions that may require farmers to take action.

## Daily Weather Observation

The field officer provides the following information.

Farm Location = Kaduna

Crop Type = Maize

Temperature (◦C) = 31

Rainfall (mm) = 18

Humidity (%) = 72

Wind Speed (km/h) = 16

Minimum Rainfall Required (mm) = 20

Maximum Safe Temperature (◦C) = 35

Irrigation System Available = True
```

 ## Task 1 - Create the Variables



```python
# Create Variable for all the information provided
# Use appropriate data types for each variable

farm_location = "Kaduna"
crop_type = "Maize"
current_temperature = 31
recorded_rainfall = 18
humidity = 72
wind_speed = 16
min_required_rainfall = 20
max_safe_temperature = 35
irrigation_system_available = True
```

## Task 2 - Calculate Rainfall Difference

```python
# Millimeters of rainfall needed to reach the minimun rainfall requirement

recorded_rainfall = 18
min_required_rainfall =20 
rainfall_difference = min_required_rainfall - recorded_rainfall
print(rainfall_difference)
print("Rainfall Difference:", rainfall_difference)
```

## Task 3 - Calculate Temperature Margin

```python
# Degrees remain before reaching the maximum safe temperature

current_temperature = 31
max_safe_temperature = 35
temperature_margin = max_safe_temperature - current_temperature
result = temperature_margin
print(result)
print("Temperature Margin:", result)
```

## Task 4 - Update Rainfall

```python
# Additional rainfall of 8mm later in the day

recorded_rainfall = 18
recorded_rainfall += 8
print("Updated Rainfall:", recorded_rainfall)
```

## Task 5 - Check if Rainfall Requirement Has Been Met

```python
# Comparison etween updated rainfall and minimum rainfall requirement

updated_rainfall = 26
min_required_rainfall = 20
print(updated_rainfall >= min_required_rainfall)
```

## Task 6 - Check if Temperature Is Within the Safe Limit

```python
# comparison between current temperature and maximum safe temperature
# Current temperature is same as recorded temperature

current_temperature = 31
max_safe_temperature = 35
print(current_temperature <= max_safe_temperature)
```

## Task 7 - Determine if Crop Conditions Are Favorable

```python
# Favorable Conditions if
# Rainfall requirement has been met
# And
# Temperature is within the safe limit

required_rainfall_met = True
temperature_safe_limit = True

result = required_rainfall_met and temperature_safe_limit
print(result)
print("Favorable Growing Conditions:" , result)
```

## Task 8 - Determine if Farmer Action Is required

```python
# Action should be taken if
# Growing conditions are not favorable
# Or
# Irrigation is not available

favorable_growing_conditions = True
irrigation_available = True

action_required = (not favorable_growing_conditions) or (not irrigation_available)
print(action_required)
print("Action should be taken:", action_required)
```

## Task 9 - Generate the weather Impact Report

```python
# Displaying of result in Professional Report

print("=" * 60)
print(" WEATHER IMPACT ANALYSIS REPORT")
print("=" * 60)
print("Farm Location:", farm_location)
print("Crop Type:", crop_type)
print("Temperature:", current_temperature, "°C")
print("Updated Rainfall:", updated_rainfall, "mm")
print("Humidity:", humidity, "%")
print("Wind Speed:", wind_speed, "km/h")
print("Rainfall Difference:", rainfall_difference, "mm")
print("Temperature Margin:", temperature_margin, "°C")
print("Minimum Rainfall Requirement:", min_required_rainfall, "mm")
print("Maximum Safe Temperature:", max_safe_temperature, "°C")
print("Rainfall Requirement Met:", required_rainfall_met)
print("Temperature Within Safe Limit:", temperature_safe_limit)
print("Irrigation Available:", irrigation_available)
print("Growing Conditions Favorable:", required_rainfall_met and temperature_safe_limit)
print("Farmer Action Required:", action_required)
```
